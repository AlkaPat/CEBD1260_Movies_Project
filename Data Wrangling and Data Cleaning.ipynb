# Fixing up Movie Dataset


```python
#Dropping Unnecessary Columns
df_movie=df_movie.drop(columns=['adult','belongs_to_collection','homepage','overview','poster_path','status','tagline',
              'video','imdb_id','spoken_languages'])
#changing name of column to match rating dataframe
df_movie = df_movie.rename(columns={'id': 'movieId'})
```

```python
#Changing Column Data types
df_movie['release_date'] = pd.to_datetime(df_movie['release_date'], errors='coerce')
print("release date is", df_movie['release_date'].dtypes)
#Changing budjet to int
df_movie['budget'] = pd.to_numeric(df_movie['budget'], errors='coerce')
print("budget is", df_movie['budget'].dtypes)
#Changing runtime to int
df_movie['runtime'] = pd.to_numeric(df_movie['runtime'], errors='coerce')
print("runtime is", df_movie['runtime'].dtypes)
#Changing name and type of movie_id to match ratings table
df_movie['movieId'] = pd.to_numeric(df_movie['movieId'], errors='coerce').fillna(0).astype(np.int64)
print("movieID is", df_movie['movieId'].dtypes)
df_movie['popularity'] = pd.to_numeric(df_movie['popularity'], errors='coerce')
print("popularity is", df_movie['popularity'].dtypes)
```

```python
"""
Need to create a list from the Nested Genres, used this function that I got from: 
    https://www.kaggle.com/rounakbanik/movie-recommender-systems
https://stackoverflow.com/questions/52064741/extracting-json-data-from-column-in-pandas-dataframe
"""


df_movie['genres'] = df_movie['genres'].fillna('[]').apply(literal_eval).apply((lambda x: [i['name'] for i in x] if isinstance(x, list) else []))
df_movie['production_companies'] = df_movie['production_companies'].fillna('[]').apply(literal_eval).apply((lambda x: [i['name'] for i in x] if isinstance(x, list) else []))

df_movie['production_countries'] = df_movie['production_countries'].fillna('[]').apply(literal_eval).apply((lambda x: [i['name'] for i in x] if isinstance(x, list) else []))
df_movie.head()
```
